```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filehandler = open("Layer Results/Fold1/id_4.obj", 'rb') 
data = pickle.load(filehandler)
filehandler.close()
```

```python
# output, target, before,
output, target, before = data
output = output.detach().cpu().numpy()
```

```python
plt.hist(output.flatten(),bins=10, alpha=0.5, label="Output")
plt.hist(target.flatten(),bins=10, alpha=0.5, label="Target")
plt.hist(before.flatten(),bins=10, alpha=0.5, label="Before")
plt.legend()
plt.show()
```

```python
plt.title("Y Axsis is the Target, X is the OOD layer")
plt.scatter(target.flatten(),output.flatten(), label="After Aligment")
plt.scatter(target.flatten(), before.flatten(), label="Before Alignment")
plt.legend()
plt.show()
```

```python
def readHospital(names, h):
    result = pd.DataFrame()
    for name in names:
        address = f"Results/{name}_fold{1}/"
        h1_fold1 = pd.read_csv(address + f"h_{h}.csv", header=None)

        address = f"Results/{name}_fold{2}/"
        h1_fold2 = pd.read_csv(address + f"h_{h}.csv", header=None)

        h1 = pd.concat((h1_fold1,h1_fold2))
        result = pd.concat((result,h1))
    return result



h1 = readHospital(["Almanak1","Almanak2","Bursa3","Bursa4"],1)

h1[0].to_numpy()
```
